# Q5 Which region would be cheaper to buy avocados from next year?
    Prices of avocados sold by regions and years will be plotted and the price will be predicted for the next year

```python
#taking necessary columns from avocado data
# WE need avocado prices and region. Also years to get regression function

df_group=data[["AveragePrice","region","year"]]
df_group
```

```python
# We groupped our taken data by region and year.
# And we calculated mean all prices in this data

df_group_reg=df_group.groupby(["region","year"]).agg(np.mean)
df_group_reg
```

```python
# We have average prices for each region in 2015, 2016, 2017 and 2018.
# We will create regression function for each region depends on year
# And then Predict avocado prices for each region in 2019 

from sklearn.linear_model import LinearRegression

years=[[2015],[2016],[2017],[2018]]

df_predict=pd.DataFrame(columns=['region','AveragePrice'])
data_types_dict = {'region': str,'AveragePrice': float} 
df_predict = df_predict.astype(data_types_dict) 

for i in df_group_reg.groupby("region"):
    predict_average_price= LinearRegression().fit(years,i[1]["AveragePrice"].values).predict([[2019]])
    new_row = {'region':i[0], 'AveragePrice':predict_average_price[0]}
    df_predict=df_predict.append(new_row,ignore_index=True)
df_predict
```

```python
# We are ready to compare avocado price for each region.
#We used plot bar and we put mean of prices for each region
#So, see which region has higher bar and which region has smallest bar

plt.figure(figsize=(15,5))  # width:15, height:5
bar=plt.bar(df_predict["region"],df_predict["AveragePrice"],align='center')
plt.xticks(df_predict.index,color='green',rotation=90, horizontalalignment='center')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
years=[[2015],[2016],[2017],[2018]]

df_predict=pd.DataFrame(columns=['region','AveragePrice'])
data_types_dict = {'region': str,'AveragePrice': float} 
df_predict = df_predict.astype(data_types_dict) 

for i in df_group_reg.groupby("region"):
    predict_average_price= DecisionTreeRegressor().fit(years,i[1]["AveragePrice"].values).predict([[2019]])
    new_row = {'region':i[0], 'AveragePrice':predict_average_price[0]}
    df_predict=df_predict.append(new_row,ignore_index=True)
df_predict
```

# Q6 Will organic or inorganic planting next year bring more income??
   Organic and inorganic avocado prices and avocado sales averages will be calculated annually.
And sales prices will be predicted for the next year.
And the sales volumes for the next year will be predicted.

Then these values will be multiplied and the average annual avocado revenues will be calculated for all years.

Finally the result is displayed with line chart.

```python
#taking necessary columns from avocado data. We need 2 dataframe.
# 1. Average prices grouped by year and type
# 2. Total Volume grouped by year and type
```

```python
# WE need avocado prices and type. Also years to get regression function

df_group_price=data[["AveragePrice","year","type"]]
df_group_price
```

```python
# We groupped our taken data by year.
# And we calculated mean all prices in this data

df_group_price=df_group_price.groupby(["type","year"]).agg(np.mean)
df_group_price
```

```python
# We have average prices for each type in 2015, 2016, 2017 and 2018.
# We will create regression function for each type depends on year
# And then Predict avocado prices for 2019 

from sklearn.linear_model import LinearRegression

years=[[2015],[2016],[2017],[2018]]

df_predict_prices=pd.DataFrame(columns=['type','AveragePrice'])
data_types_dict = {'type': str,'AveragePrice': float} 
df_predict_prices = df_predict_prices.astype(data_types_dict) 

for i in df_group_price.groupby("type"):
    predict_average_price= LinearRegression().fit(years,i[1]["AveragePrice"].values).predict([[2019]])
    new_row = {'type':i[0], 'AveragePrice':predict_average_price[0]}
    df_predict_prices=df_predict_prices.append(new_row,ignore_index=True)
print("Average avocado price in 2019 according to type")
df_predict_prices
```

```python
from sklearn.tree import DecisionTreeRegressor
years=[[2015],[2016],[2017],[2018]]

df_predict_prices=pd.DataFrame(columns=['type','AveragePrice'])
data_types_dict = {'type': str,'AveragePrice': float} 
df_predict_prices = df_predict_prices.astype(data_types_dict) 

for i in df_group_price.groupby("type"):
    predict_average_price= DecisionTreeRegressor().fit(years,i[1]["AveragePrice"].values).predict([[2019]])
    new_row = {'type':i[0], 'AveragePrice':predict_average_price[0]}
    df_predict_prices=df_predict_prices.append(new_row,ignore_index=True)
print("Average avocado price in 2019 according to type")
df_predict_prices
```

```python
# we need to do same things for total volume
# taking total volume, year and type columns from data

df_group_volume=data[["Total Volume","year","type"]]
df_group_volume
```

```python
# We groupped our taken data by year and type.
# And we calculated mean all volumes in this data

df_group_volume=df_group_volume.groupby(["type","year"]).agg(np.mean)
df_group_volume
```

```python
# We have average prices for each type in 2015, 2016, 2017 and 2018.
# We will create regression function for each type depends on year
# And then Predict avocado prices for each type in 2019 

from sklearn.linear_model import LinearRegression

years=[[2015],[2016],[2017],[2018]]

df_predict_volumes=pd.DataFrame(columns=['type','Total Volume'])
data_types_dict = {'type': str,'Total Volume': float} 
df_predict_volumes = df_predict_volumes.astype(data_types_dict) 

for i in df_group_volume.groupby("type"):
    predict_total_volume= LinearRegression().fit(years,i[1]["Total Volume"].values).predict([[2019]])
    new_row = {'type':i[0], 'Total Volume':predict_total_volume[0]}
    df_predict_volumes=df_predict_volumes.append(new_row,ignore_index=True)
print("Total Volume of avocado averages in 2019 according to type")
df_predict_volumes
```

```python
# adding predicted values to total volume dataframe

value1= df_predict_volumes["Total Volume"][0]
df_group_volume.loc[('conventional', 2019), :] = (value1)
value2= df_predict_volumes["Total Volume"][1]
df_group_volume.loc[('organic', 2019), :] = (value2)
df_group_volume=df_group_volume.sort_index()
df_group_volume
```

```python
# adding predicted values to avocado prices dataframe

value1= df_predict_prices["AveragePrice"][0]
df_group_price.loc[('conventional', 2019), :] = (value1)
value2= df_predict_prices["AveragePrice"][1]
df_group_price.loc[('organic', 2019), :] = (value2)
df_group_price=df_group_price.sort_index()
df_group_price
```

```python
# Splitting types and have organic, conventional data frames for price and volume
# then multiple averageprice x total volume values to calculate incomes for each year

df_convential_price=df_group_price.groupby("type").get_group("conventional").droplevel('type')
df_organic_price=df_group_price.groupby("type").get_group("organic").droplevel('type')

df_convential_volume=df_group_volume.groupby("type").get_group("conventional").droplevel('type')
df_organic_volume=df_group_volume.groupby("type").get_group("organic").droplevel('type')

df_result_conventional=df_convential_price.values*df_convential_volume.values
df_result_organic=df_organic_price.values*df_organic_volume.values
```

```python
# show that we have with line graph

years=[2015,2016,2017,2018,2019]
plt.plot( years,df_result_conventional, 'rs--',label="conventional")
plt.plot(years,df_result_organic, 'bs:',label="organic")
plt.xlabel('Type')
plt.ylabel('Average annual avocado revenues in 2019')
plt.legend()
plt.show()
```

```python
#We see conventional avocados have more income every year. and it will be same in 2019.
# If we want to print comparing for 2019 year

if df_result_conventional[4]>df_result_organic[4]:
    print(" If you farm Conventional avocados you will get more income")
else:
     print(" If you farm Organic avocados you will get more income")
```

# Classification

```python
#Importing the libraries we will use
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import statsmodels.api as sm 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

data.head()
```

```python
#Label encoder assigns a numerical value to each categorical data.
type=data["type"].values
le=LabelEncoder()
type = le.fit_transform(type)
type=pd.DataFrame(type,columns=["type"])

region = data["region"].values
region = le.fit_transform(region)
region =pd.DataFrame(region ,columns=["region"])

data=data.drop(["Date","region","type","year"],axis=1) #year 
data = pd.concat([data,region,type],axis=1)
data.head()
```

```python
y = data["type"].values
x = data.drop("type",axis=1).values
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=0)
```

```python
#We will use Min Max Scaler
ms=MinMaxScaler()

x_train = ms.fit_transform(x_train)
x_test = ms.transform(x_test)
```

```python
scores = []
```

## Logistic Regression

```python
print("Logistic Regression\n")
lr=LogisticRegression(random_state=0)
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)

cm_logr=confusion_matrix(y_test,y_pred)
score = accuracy_score(y_test,y_pred)
scores.append(["LogisticRegression",score,"","","",""])

print("\tClassification Report")
print(classification_report(y_test,y_pred))
print("\nScore:",score)
print("Consufion Matrix")
print(cm_logr)
```

## K Nearest Neighbors

```python
for i in [1,3,5,10]:
    knn = KNeighborsClassifier(n_neighbors=3,metric= "euclidean")
    knn.fit(x_train,y_train)
    y_pred = knn.predict(x_test)
    cm_knn=confusion_matrix(y_pred,y_test)
    score = accuracy_score(y_test,y_pred)
    scores.append(["KNeighborsClassifier",score,"",i,"euclidean",""])
    print("K Nearest Neighbors\n")
    print("K:",i)
    print("Classification Report")
    print(classification_report(y_test,y_pred))
    print("\nScore:",score)
    print("Consufion Matrix")
    print(cm_knn)
   
```

```python

for i in [1,3,5,10]:
    knn = KNeighborsClassifier(n_neighbors=3,metric= "minkowski")
    knn.fit(x_train,y_train)
    y_pred = knn.predict(x_test)
    cm_knn=confusion_matrix(y_pred,y_test)
    score = accuracy_score(y_test,y_pred)
    scores.append(["KNeighborsClassifier",score,"",i,"minkowski",""])
    
    print("K Nearest Neighbors\n")
    print("K:",i)
    print("Classification Report")
    print(classification_report(y_test,y_pred))
    print("\nScore:",accuracy_score(y_test,y_pred))
    print("Consufion Matrix")
    print(cm_knn)
```

## Support Vector Machine

```python
for kernel in ["linear", "poly", "rbf", "sigmoid"]:
    svc = SVC(kernel = kernel,random_state=0)
    svc.fit(x_train,y_train)
    y_pred = svc.predict(x_test)
    cm_svc = confusion_matrix(y_pred,y_test)
    score = accuracy_score(y_test,y_pred)
    scores.append(["SupportVectorMachine",score,"","","",kernel])
    
    print("Support Vector Machine\n")
    print("\n"+kernel)
    print("Classification Report")
    print(classification_report(y_test,y_pred))
    print("\nScore:",accuracy_score(y_test,y_pred))
    print("Confusion Matrix")
    print(cm_svc)
    
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
y_pred = gnb.predict(x_test)
cm_gnb = confusion_matrix(y_test,y_pred)
score = accuracy_score(y_test,y_pred)
scores.append(["GaussianNaiveBayes",score,"","","",""])

print("Gaussian Naive Bayes")
print("Classification Report")
print(classification_report(y_test,y_pred))
print("\nScore:",accuracy_score(y_test,y_pred))
print("Confusion Matrix")
print(cm_gnb)
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
mnb.fit(x_train,y_train)
y_pred = mnb.predict(x_test)

cm_mnb = confusion_matrix(y_test,y_pred)
score = accuracy_score(y_test,y_pred)
scores.append(["MultinomialNaiveBayes",score,"","","",""])

print("Gaussian Naive Bayes")
print("Classification Report")
print(classification_report(y_test,y_pred))
print("\nScore:",accuracy_score(y_test,y_pred))
print("Confusion Matrix")
print(cm_mnb)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion = 'gini',random_state=0)
dtc.fit(x_train,y_train)
y_pred = dtc.predict(x_test)

cm_dtc = confusion_matrix(y_test,y_pred)
score = accuracy_score(y_test,y_pred)
scores.append(["DecisionTreeClassifier",score,"gini","","",""])

print("Classification Report")
print(classification_report(y_test,y_pred))
print("\nScore:",accuracy_score(y_test,y_pred))
print("Confusion Matrix")
print(cm_dtc)
```

```python
dtc = DecisionTreeClassifier(criterion = 'entropy',random_state=0)
dtc.fit(x_train,y_train)
y_pred = dtc.predict(x_test)

cm_dtc = confusion_matrix(y_test,y_pred)
score = accuracy_score(y_test,y_pred)
scores.append(["DecisionTreeClassifier",score,"entropy","","",""])

print("Classification Report")
print(classification_report(y_test,y_pred))
print("\nScore:",accuracy_score(y_test,y_pred))
print("Confusion Matrix")
print(cm_dtc)
```

## Random Forest  

```python
from sklearn.ensemble import RandomForestClassifier
for n_estimators in [100,150,200]:
    rfc = RandomForestClassifier(n_estimators=n_estimators, criterion = 'gini',random_state=0) 
    rfc.fit(x_train,y_train) #n_estimators: The number of trees in the forest.
    y_pred = rfc.predict(x_test)
    cm_rfc = confusion_matrix(y_test,y_pred)
    score = accuracy_score(y_test,y_pred)
    scores.append(["RandomForestClassifier",score,"gini",n_estimators,"",""])

    print("RandomForestClassifier\n")
    print("Classification Report")
    print(classification_report(y_test,y_pred))
    print("\nScore:",accuracy_score(y_test,y_pred))
    print("Confusion Matrix Gini")
    print(cm_rfc)
    
```

```python
for n_estimators in [100,150,200]:
    rfc = RandomForestClassifier(n_estimators=n_estimators, criterion = 'entropy',random_state=0) 
    rfc.fit(x_train,y_train)
    y_pred = rfc.predict(x_test)
    cm_rfc = confusion_matrix(y_test,y_pred)
    score = accuracy_score(y_test,y_pred)
    scores.append(["RandomForestClassifier",score,"entropy",n_estimators,"",""])
    print("RandomForestClassifier\n")
    print("Classification Report")
    print(classification_report(y_test,y_pred))
    print("\nScore:",accuracy_score(y_test,y_pred))
    print("Confusion Matrix Entropy")
    print(cm_rfc)
```
